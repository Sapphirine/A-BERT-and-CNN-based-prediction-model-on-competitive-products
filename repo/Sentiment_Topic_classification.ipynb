```python
y = table['Apple']
#table.drop(['Apple', 'Username'], inplace=True, axis=1)
table.drop(['Apple'], inplace=True, axis=1)
```

```python
X = table
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
X_test
```

```python
model = Sequential()
# First we add an embedding layer 
model.add(Embedding(vocab_size,embedding_dims,input_length=max_len)) 
# Adding a 1D convolutional Layer
model.add(Conv1D(filters, ker_size, padding='valid', activation='relu'))
# Max Pooling the Convolutions
model.add(MaxPooling1D())
# Again Computing the Convolutions
model.add(Flatten())
model.add(Dense(hidden_dims, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train, batch_size=batch_size, epochs=100, validation_data=(X_test, y_test))
```

```python
model.evaluate(X_test,y_test)[1]*100
```
